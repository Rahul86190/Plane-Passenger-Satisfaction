# Start data preparation

```python
df
```

Data visualixation = Changing necessary values form catbolical to integer

ONEHOTENCODING  by pandas



```python
pd.get_dummies(df)
```

```python
pd.get_dummies(df).astype(int)
```

```python
df2=pd.get_dummies(df,drop_first="if_binary").astype(int)
df2.head()
```

```python
x=df2.drop("satisfaction_satisfied",axis=1)      #independent variable
```

```python
y=df2[["satisfaction_satisfied"]] #target variable
```

training 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
```

train algo

```python
lgr.fit(x_train,y_train)
```

```python
lgr.score(x_train,y_train)*100
```

```python
lgr.score(x_test,y_test)*100
```

```python
predeiction=lgr.predict(x_test)
```

```python
y_test
```

```python
y_test["predeiction"]=predeiction
```

## Model evaluation test

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# confusion_matrix(actual,predeiction)
matrix=confusion_matrix(y_test["satisfaction_satisfied"],y_test["predeiction"])
matrix
```

```python
sns.heatmap(matrix,annot=True,fmt="d")  # d -> decimal
plt.title("confusion_matrix")
plt.axis("off")
plt.show()
```

```python
import os,joblib
```

```python
os.makedirs("model",exist_ok=True)
joblib.dump(lgr,'./model/logisticregres.lb')
```

```python
print(classification_report(y_test["satisfaction_satisfied"],y_test["predeiction"]))
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
```

```python
dtc=DecisionTreeClassifier()
rdf=RandomForestClassifier()
```

```python
dtc.fit(x_train,y_train)
rdf.fit(x_train,y_train)
print("your algo is trained")
```

```python
print("Decision Tree Score")
print("your training score:",dtc.score(x_train,y_train)*100)
print("testing score:",dtc.score(x_test,y_test['satisfaction_satisfied'])*100)

print("Random ForestScore")
print("your training score:",rdf.score(x_train,y_train)*100)
print("testing score:",rdf.score(x_test,y_test['satisfaction_satisfied'])*100)
```
